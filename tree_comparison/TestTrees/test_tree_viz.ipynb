```python
from neuron_morphology.swc_io import morphology_from_swc
from morph_utils.visuals import basic_morph_plot
import glob
import matplotlib.pyplot as plt
```

```python
root = r'\\allen\programs\celltypes\workgroups\mousecelltypes\SarahWB\github_projects\tree_comparison\tree_comparison\TestTrees\*.swc'
for file in glob.glob(root):
    fig, ax = plt.subplots(nrows=1, ncols=1)
    nrn = morphology_from_swc(file)
    swc_name = file.rsplit('\\',1)[1]
    if swc_name[0] == '1': 
        scat = True
        lw = 0.01
        ds = 0.01
    else: 
        scat = True
        lw = 0.5
        ds = None
    basic_morph_plot(nrn, ax, scatter=scat, scatter_roots=False, scatter_soma=False, soma_dot_size=0, plot_soma=False, line_w=lw, dotsize=ds)
    ax.set_title(file.rsplit('\\',1)[1])
    fig.savefig(file.replace('.swc', '.pdf'))
```
